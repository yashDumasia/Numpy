# Advance Operation With Business Examples :

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure : [restaurant_id, 2021 , 2022 , 2023 , 2024]
sale_data = np.array([      
    [1, 150000, 180000, 125000, 200000], # Pizza Hub
    [2, 155000, 160000, 225000, 196000], # Burger King
    [3, 152500, 200000, 220000, 300000], # Burger Point
    [4, 170000, 118000, 180000, 200000], # Chai Point 
    [5, 189700, 200000, 225000, 300000], # KFC
])
print("=== Zomato Sale Data ===")
print("\n")
print("Shape of Sale Data :", sale_data.shape)
print("Sale of First 3 Restaurant :\n",sale_data[0:3,:])
```

```python
# Total Sales per year
yearly_total = np.sum(sale_data[:,1:],axis = 0)
print("Total Sales per year :",yearly_total)

print("\n")

# Minimum Sales per restaurant
min_sales = np.min(sale_data[:,1:],axis = 1)
print("Minimum Sales per Restaurant :",min_sales)

print("\n")

# Maximum Sales per restaurant
max_sales = np.max(sale_data[:,1:],axis = 1)
print("Maximum Sales per Restaurant :",max_sales)

print("\n")

# Average Sales per restaurant
ave_sales = np.average(sale_data[:,1:],axis = 1)
print("Average Sales per Restaurant :",ave_sales)
```

```python
cumsum = np.cumsum(sale_data[:,1:],axis = 1)
#cumsum = [sale(2021)   sale(2021+2022)   sale(2021+2022+2023) .....]
print("Cumsum :\n",cumsum)


plt.figure(figsize = (8,5))
plt.plot(np.mean(cumsum,axis = 0))
plt.title("Aerage Cumulative sales across all restaurant. ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector Addition :",vector1+vector2)
print("\n")
print("Vector Multiplication :",vector1*vector2)
print("\n")
print("Vector Dot Product :",np.dot(vector1,vector2))
print("\n")
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between Vectors :",angle)
```
